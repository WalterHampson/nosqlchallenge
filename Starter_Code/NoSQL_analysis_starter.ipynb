### 1. Which establishments have a hygiene score equal to 20?

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Define the latitude and longitude values for London
london_latitude = 51.5074
london_longitude = -0.1278

# Convert latitude and longitude strings from the database to floats
latitude_db = float("51.083812")
longitude_db = float("1.195625")

# Define the degree search radius
degree_search = 0.01

# Define the query
query = {
    "geocode.latitude": {"$gte": str(latitude_db - degree_search), "$lte": str(latitude_db + degree_search)},
    "geocode.longitude": {"$gte": str(longitude_db - degree_search), "$lte": str(longitude_db + degree_search)},
    "RatingValue": "5"
}

# Define the sort order
sort = [("scores.Hygiene", 1)]  # Sort by hygiene score in ascending order

# Define the limit for the number of results
limit = 5

# Print the results
print("Top 5 establishments near London with RatingValue of 5, sorted by hygiene score:")
for doc in establishments.find(query).sort(sort).limit(limit):
    pprint(doc)
```

```python
# Execute the query and store the result in a variable
cursor = establishments.find(query).sort(sort).limit(limit)

# Convert the cursor to a list of dictionaries
result_list = list(cursor)

# Convert the result to a Pandas DataFrame
df = pd.DataFrame(result_list)

# Print the DataFrame
print("DataFrame:")
print(df)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Define the pipeline stages
pipeline = [
    # Match establishments with a hygiene score of 0
    {"$match": {"scores.Hygiene": 0}},
    # Group the matches by Local Authority
    {"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}},
    # Sort the matches from highest to lowest
    {"$sort": {"count": -1}}
]

# Execute the pipeline
result = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print("Number of documents in the result:", len(result))

# Print the first 10 results
print("First 10 results:")
for i, doc in enumerate(result[:10], 1):
    print(f"{i}. Local Authority: {doc['_id']}, Count: {doc['count']}")
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(result)

# Display the number of rows in the DataFrame
print("Number of rows in the DataFrame:", len(df))

# Display the first 10 rows of the DataFrame
print("First 10 rows of the DataFrame:")
print(df.head(10))
```
